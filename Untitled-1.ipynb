```python
model_root = 'models'
my_net = torch.load(os.path.join(
        model_root, 'mnist_mnistm_model_epoch_current.pth'
    ))
my_net = my_net.eval()

my_net = my_net.cuda(3)

_,dataloader,_ = load_data("","","",batch_size = 32)
data_target_iter = iter(dataloader)
len_dataloader = len(dataloader)
i = 0
while i < len_dataloader:
    
    data = data_target_iter.next()
    t_img, t_label = data
    t_img = t_img.cuda(3)
    output = my_net(input_data = t_img ,alpha =0)
    if i == 0:
        res = output.data.cpu()
    else:
        res = torch.cat((res , output.data.cpu()),dim=0)
    i = i+1
torch.save(res , "cwt.torch")
```

```python
import pywt
import numpy as np

# 创建信号
x = np.linspace(0, 1, 1024)
signal = np.sin(250 * np.pi * x) + np.sin(500 * np.pi * x)

# 进行小波变换
wavelet = 'db2' # 选择小波基函数
levels = 4     # 分解层数
coeffs = pywt.wavedec(signal, wavelet, level=levels)
# 可视化小波变换结果
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 8))

# 绘制原始信号
ax.plot(x, signal, label='Original Signal', color='blue')

# 绘制小波变换后的信号
for i in range(levels):
    ax.plot(x[:len(coeffs[i+1])], coeffs[i+1], label=f'Detail {i+1}', alpha=0.5)
ax.plot(x[:len(coeffs[0])], coeffs[0], label='Approximation', color='red')

# 添加标签和图例
ax.set_title('Wavelet transform of a signal')
ax.set_xlabel('Time')
ax.legend()

# 显示图像
plt.show()
```

```python
from pywt import scale2frequency
wavename = 'gaus1'

totalscal = 129     # totalscal是对信号进行小波变换时所用尺度序列的长度(通常需要预先设定好)
fc = pywt.central_frequency(wavename)  # 计算小波函数的中心频率
cparam = 2 * fc * totalscal  # 常数c
scales = cparam/np.arange(totalscal, 1, -1) 
x = np.arange(2048)
y = np.sin(2*np.pi*x/32)

coef, freqs=pywt.cwt(y,scales,'gaus1')
plt.matshow(coef) # doctest: +SKIP
plt.show() 
```
